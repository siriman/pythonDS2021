```python
country_h_co2= data_2011.sort_values(by='CO2 Per Capita (metric tons)', ascending=False)
```

```python
top_10 = country_h_co2[:5]
```

```python
import seaborn as sns
```

```python
sns.barplot(x=top_10['Country Name'], y=top_10['CO2 Per Capita (metric tons)'])
```
