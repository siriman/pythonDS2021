#### Decorer votre plot

You can then detail properly:
- le titre de la figure: `plt.title('Chart Title')`
- les labels des axes: avec `plt.xlabel("X-title")` et `plt.ylabel("Y-title")`

Comme vous le verrez dans la documentation, matplotlib propose une vari√©t√© de param√®tres dans ces fonctions pour personnaliser vos graphiques comme vous le souhaitez.

Apr√®s tout cela, vous appelez la fonction `plt.show ()` pour dessiner la figure.

````python
plt.xlabel('X-axis (in units)', fontsize='14')
plt.ylabel('Y-axis (in units)', fontsize='14')

plt.title('My wonderful chart', 
          fontweight='bold', color = 'black', fontsize='16', alpha=0.8)

plt.show()
````

#### plt.figure

Lors de la cr√©ation d'un graphique avec `pyplot`, il cr√©e automatiquement une figure (conteneur pour vos graphiques).

Mais vous pouvez le cr√©er vous-m√™me, afin par exemple de sp√©cifier la taille de votre figurine avec `figsize = (largeur, hauteur)` avec des dimensions en pouces.

````python
plt.figure(figsize=(12,6))

plt.xlabel('X-axis (in units)', fontsize='14')
plt.ylabel('Y-axis (in units)', fontsize='14')

plt.title('My wonderful chart', 
          fontweight='bold', color = 'black', fontsize='16', alpha=0.8)

plt.show()
````

#### plt.subplots

Il est possible de cr√©er plus d'un graphique dans une figure en utilisant la fonction `plt.subplots`.

Vous pouvez utiliser `plt.tight_layout` pour organiser automatiquement vos trac√©s (et √©viter les superpositions par exemple)

````python
plt.subplots(nrows=4, ncols=4)

plt.tight_layout()

plt.show()
````

#### plt.hist

Vous pouvez facilement dessiner des histogrammes avec `pyplot` en utilisant la fonction` plt.hist`.

Cela vous permet de visualiser la **distribution** de vos donn√©es dans diff√©rents **bins** (ou barres ou buckets) de m√™me taille.

Les bins repr√©sentent une plage de valeurs identique, la hauteur de la barre correspond au nombre de valeurs contenues dans cette plage.

````python
import numpy as np
mu = 200
sigma = 25
x = np.random.normal(mu, sigma, size=1000)

plt.figure(figsize=(10,5))

plt.hist(x, bins=20, facecolor='blue', alpha=0.5)

plt.title('Histogram of Normal Distribution (random draw)', fontweight='bold')
plt.show()
````

#### plt.scatter

Vous pouvez visualiser les points de donn√©es sur un axe X vs Y en utilisant `plt.scatter`

C'est tr√®s utile lors de la recherche de **corr√©lations** entre 2 variables par exemple.

Cette fonction trace un nuage de points des observations. Il affiche g√©n√©ralement les valeurs de deux variables pour un ensemble de donn√©es. Vous pouvez √©galement afficher plus de variables en colorant les points ou en variant la taille.

````python
x = range(1,101)
y = x + np.random.randn(100) * 15

plt.figure(figsize=(10,5))

plt.scatter(x=x, 
            y=y, 
            marker='o',
            alpha=0.5)

plt.show()
````

````python
x = np.random.rand(100) - 0.5
y = x + np.random.rand(100)
z = x + x * np.random.rand(100)

plt.figure(figsize=(10,5))

plt.scatter(x, y, s=z * 1000, alpha=0.5)

plt.show()
````

````python
import seaborn as sns
import pandas as pd
df = sns.load_dataset('iris')
df['species'] = pd.Categorical(df['species'])

plt.figure(figsize=(10,5))

plt.scatter(x=df['sepal_length'],
            y=df['sepal_width'],
            c=df['species'].cat.codes,
            cmap="Set1",
            alpha=0.8)

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width') 
plt.show()
````

#### plt.plot

Un des graphiques les plus classiques: `plt.plot`. Tracez des lignes entre les points de donn√©es. Particuli√®rement utile avec les s√©ries temporelles.

````python
plt.figure(figsize=(10,5))

x = range(1900, 2000)
y = range(100) + np.random.randn(100) * 15

plt.plot(x, y,
         marker='x',
         linestyle='--',
         color='mediumvioletred',
         alpha=0.5)

plt.title("Evolution of xxx during 21th century")
plt.show()
````

#### plt.bar

Vous pouvez cr√©er un **diagramme en barre** avec `plt.bar`.

Il est principalement utilis√© pour comparer des donn√©es avec quelques √©l√©ments de quelques cat√©gories.

````python
plt.figure(figsize=(10, 5))

bars = ['Q1 2018', 'Q2 2018', 'Q3 2018', 'Q4 2018']
sales = [12, 14, 25, 59]

plt.bar(x=bars,
        height=sales)

plt.ylabel("Sales (in M‚Ç¨)")
plt.title("Sales have skyrocketed Month-Over-Month (MoM)!", 
          fontweight='bold', fontsize='14', alpha=0.8)
plt.show()
````

#### plt.stackplot

Si vous souhaitez observer la composition de certaines donn√©es et leur √©volution **relative** dans le temps. Vous pouvez utiliser la fonction `plt.stackplot`.

````python
data = pd.DataFrame({'group_A':[1,4,6,8,9],
                     'group_B':[2,24,7,10,12],
                     'group_C':[2,8,5,10,6], },
                    index=range(1,6)
                   )
data_perc = data.divide(data.sum(axis=1), axis=0)

plt.figure(figsize=(10, 5))

plt.stackplot(range(1,6),
              data_perc["group_A"], 
              data_perc["group_B"], 
              data_perc["group_C"],
              labels=['A','B','C'])

plt.legend(loc='upper left')
plt.margins(0,0)
plt.title('Evolution of market share over time')
plt.show()
````

#### plt.subplot - digging deeper

D√©monstrons maintenant bri√®vement comment pr√©senter plusieurs graphiques en m√™me temps avec `plt.subplot`

Cela fonctionne comme ceci: `plt.subplot (nrows, ncolumns, index)`



````python
x = range(1,101)
y = np.random.randn(100) * 15 + range(1,101)
z = (np.random.randn(100) * 15 + range(1,101)) * 2
````

````python
plt.figure(figsize=(10, 8))

plt.subplot(211)
plt.plot(x, y, marker='o', alpha=0.4)
 
plt.subplot(212)
plt.plot(x, y, linestyle='none', marker='o', color="orange", alpha=0.5)

plt.show()
````

### Seaborn

#### Boxplot - To show distributions with respect to categories.

Une boxplot (ou bo√Æte √† moustaches) montre la distribution des donn√©es quantitatives d'une mani√®re qui facilite les comparaisons entre les variables ou entre les niveaux d'une variable cat√©gorielle.

Le box plot permet un examen graphique rapide de vos donn√©es:
- Les box plot montrent les **[quartiles] (https://en.wikipedia.org/wiki/Quartile)**
- La bande √† l'int√©rieur de la bo√Æte est le **deuxi√®me quartile (la m√©diane)**
- Les extr√©mit√©s des moustaches repr√©sentent **le minimum et le maximum de toutes les donn√©es** (la plupart du temps - parfois les extr√©mit√©s des box peuvent repr√©senter d'autres valeurs)

<img src="images/boxplot.png" width=400px />

> üìö **Resources** : Pour mieux comprendre le boxplot chart, lire [Wikipedia - Box Plot](https://en.wikipedia.org/wiki/Box_plot)

````python
import seaborn as sns
````

````python
plt.figure(figsize=(12, 6))

df = sns.load_dataset("tips")

sns.boxplot(x="total_bill", data=df)

plt.title('Boxplot of total_bill variable')
plt.show()
````

#### Histogram plot - Pour montrer la distribution univari√©e des observations

La fonction seaborn `sns.distplot` combine la fonction matplotlib `hist` (avec le calcul automatique d'une bonne taille de bin par d√©faut) avec la fonction seaborn `kdeplot ()` (qui trace une estimation de densit√© lisse).

````python
plt.figure(figsize=(12, 6))

df = sns.load_dataset("iris")

sns.distplot(df["sepal_length"])

plt.title('Distribution of Sepal Length')
plt.show()
````

#### Scatter plot - Pour tracer une collection de points de donn√©es

La fonction `sns.scatterplot` dessine un nuage de points des observations. Il affiche g√©n√©ralement les valeurs de deux variables pour un ensemble de donn√©es. Vous pouvez √©galement afficher plus de variables en colorant les points ou en variant la taille.

````python
plt.figure(figsize=(12, 6))

sns.scatterplot(x="sepal_length", y="sepal_width", data=df, hue="species")

plt.show()
````

#### Pair plot - a magic function üé©

`sns.pairplot` est assez magique, vous l'appelez sur votre jeu de donn√©es et il dessine **des relations par paires dans un jeu de donn√©es**.

√áa signifie:
- chaque variable dans les donn√©es sera partag√©e sur l'axe des y sur une seule ligne et sur l'axe des x sur une seule colonne - `scatter_plot`
- Les axes diagonaux sont trait√©s diff√©remment, dessinant un trac√© pour montrer la distribution univari√©e des donn√©es pour la variable dans cette colonne - `kde_plot`

````python
sns.set(style="ticks")

sns.pairplot(df, hue="species")

plt.show()
````

#### Rel plot - Facing scatter plots

Une autre fonction seaborn utile: `sns.relplot` pour tracer des nuages de points de 2 faces partageant le m√™me axe X-Y et la m√™me l√©gende mais avec une colonne variable d√©finie dans`col`

````python
sns.set()

df_tips = sns.load_dataset("tips")

sns.relplot(x="total_bill",
            y="tip",
            col="time",
            palette="YlGnBu",
            hue="smoker",
            style="smoker",
            size="size",
            data=df_tips)

plt.show()
````

### III.1.C. Bokeh

[`bokeh`](https://bokeh.pydata.org/en/latest/) est une autre biblioth√®que de visualisation Python avec des fonctionnalit√©s d'interactivit√©.

Par exemple: [Iris visualization](https://bokeh.pydata.org/en/latest/docs/gallery/iris.html) üå∫

````python
from bokeh.plotting import figure, show, output_file
from bokeh.sampledata.iris import flowers

colormap = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
colors = [colormap[x] for x in flowers['species']]

p = figure(title = "Iris Morphology")
p.xaxis.axis_label = 'Petal Length'
p.yaxis.axis_label = 'Petal Width'

p.circle(flowers["petal_length"], flowers["petal_width"],
         color=colors, fill_alpha=0.2, size=10)

show(p)
````
